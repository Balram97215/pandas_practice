### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
chip = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
chip.head()
```

### Step 4. How many products cost more than $10.00?

```python
chip['item_price_clean'] = chip['item_price'].str.replace('$', '') #remove the dollar sign
chip['item_price_clean'] = chip['item_price_clean'].astype(float) # convert the column into float

# Now lets get the desired output

chip[chip['item_price_clean']>10].sort_values(by='item_price_clean',ascending=False).head()
```

### Step 5. What is the price of each item?
###### print a data frame with only three columns item_name choice_description and product_price

```python
columns = chip.loc[:, ['item_name', 'choice_description', 'item_price_clean']]
columns
```

### Step 6. Sort by the name of the item

```python
columns.sort_values(by='item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
max_price_items = chip[chip['item_price_clean'] == chip['item_price_clean'].max()]
# If you want the sum of quantities for these most expensive items:
quantity_of_most_expensive_item = max_price_items['quantity'].sum()
quantity_of_most_expensive_item
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie_bowl = chip[chip['item_name']== 'Veggie Salad Bowl']
time_ordered = veggie_bowl['order_id'].count()
time_ordered
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
canned_soda = chip[chip['item_name'] == 'Canned Soda']
total_order = canned_soda[canned_soda['quantity']>1].count()
total_order
```
